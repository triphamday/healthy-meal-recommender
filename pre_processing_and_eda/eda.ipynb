```python
def create_boxplot(df, column_name, color):
    fig = go.Figure()
    fig.add_trace(go.Box(
        x=df[column_name], # Use x instead of y for horizontal plot
        name=column_name,
        boxmean=True,
        orientation='h', # Set orientation to horizontal
        marker=dict(
            symbol='circle',  # Circle symbol for outliers
            size=8,  # Size of the outliers
            color=f'rgba{(*color[:3], 1)}',  # Marker color
            line=dict(color=f'rgba{(*color[:3], 1)}', width=1)  # Line color and width
        ),
        line_color=f'rgba{(*color[:3], 1)}',  # Line color
        fillcolor=f'rgba{(*color[:3], 0.5)}'  # Lighter fill color with transparency 0.5
    ))
    fig.update_layout(
        title=f'{column_name}',
        xaxis_title='Value', # Update x-axis title for horizontal plot
        yaxis_title=column_name, # Update y-axis title for horizontal plot
        showlegend=False # Hide legend for individual box plots
    )
    fig.show()
```
